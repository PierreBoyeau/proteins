```python
import pandas as pd
```

```python
feature_importances = pd.read_csv('export/feature_importances.csv', sep='\t', header=None, names=['name', 'value'])
feature_importances.loc[:, 'type'] = feature_importances['name'].str.split(pat='_', expand=True).iloc[:, 0]
feature_importances.loc[:, 'abs_value'] = feature_importances.value.abs()
feature_importances.head()
```

```python
(feature_importances.groupby('type').abs_value.mean().sort_values(ascending=False))
```

```python
(feature_importances.groupby('type').abs_value.std().sort_values(ascending=False))
```

```python
feature_importances.sort_values(by='abs_value', ascending=False)
```
